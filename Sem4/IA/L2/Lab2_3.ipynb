```python
import nltk
nltk.download('punkt')
import rowordnet as rwn

file1 = open("data/texts.txt", "r")
text = file1.read()
wn = rwn.RoWordNet()
```

```python
#numarul de propozitii din text;
len(nltk.sent_tokenize(text))
```

```python
#numarul de cuvinte din text
def deleter(a,b):
    for x in b:
        try:
            a.remove(x)
        except ValueError:
            pass
    return a

cuvinte = nltk.word_tokenize(text)
deleter(cuvinte,[",",".",":",";","!","”"])
len(cuvinte)
```

```python
#numarul de cuvinte diferite din text
cuvinte_unice_fara_semne = list(set(nltk.word_tokenize(text))-set([",",".",":",";","!","”"]))
len(cuvinte_unice_fara_semne)
```

```python
#cel mai scurt si cel mai lung cuvant (cuvinte)
minim = len(min(cuvinte_unice_fara_semne,key=len))
maxim = len(max(cuvinte_unice_fara_semne,key = len))

print([word for word in cuvinte_unice_fara_semne if len(word)==minim])
cuvinte_max = [word for word in cuvinte_unice_fara_semne if len(word)==maxim]
print(cuvinte_max)
```

```python
#textul fara diacritice
text = text.replace("ă","a").replace("â","a").replace("î","i").replace("ș","s").replace("ț","t").replace("Ă","A").replace("Â","A").replace("Î","I").replace("Ș","S").replace("Ț","T")
text
```

```python
cuvinte_max = ["laborator"]
for word in cuvinte_max:
  synset_ids = wn.synsets(literal=word)
  for synset_id in synset_ids:
    synset_object = wn.synset(synset_id)
    print("Print its literals (synonym words): {}".format(synset_object.literals))
```
