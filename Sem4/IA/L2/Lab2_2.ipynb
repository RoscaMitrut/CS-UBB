```python
import matplotlib.pyplot as plt
from PIL import Image,ImageFilter,ImageOps
import os, os.path

imgs = []
path = "data/pictures"
valid_images = [".webp",".jpg",".png"]

for f in os.listdir(path):
    ext = os.path.splitext(f)[1]
    if ext.lower() not in valid_images:
        continue
    imgs.append(Image.open(os.path.join(path,f)))
```

```python
#sa se vizualizeze una din imagini
display(imgs[0])
```

```python
#daca imaginile nu aceeasi dimensiune, sa se redimensioneze toate la 128 x 128 pixeli si sa se vizualizeze imaginile intr-un cadru tabelar.
resized_imgs = []
for image in imgs:
  resized_imgs.append(image.resize((128,128)))

fig = plt.figure()
rows=3
columns=4
i=0
for image in resized_imgs:
  if(image.mode == "L"):
    i+=1
    fig.add_subplot(rows,columns,i)
    plt.imshow(image, cmap='gray')
    plt.axis("off")
  else:
    i+=1
    fig.add_subplot(rows,columns,i)
    plt.imshow(image)
    plt.axis("off")
```

```python
#sa se transforme imaginile in format gray-levels si sa se vizualizeze
gray_imgs = []
for image in imgs:
  gray_imgs.append(ImageOps.grayscale(image))

fig2 = plt.figure()
rows=3
columns=4
i=0
for image in gray_imgs:
  i+=1
  fig2.add_subplot(rows,columns,i)
  plt.imshow(image, cmap='gray')
  plt.axis("off")
```

```python
#sa se blureze o imagine si sa se afiseze in format "before-after"
blurred = imgs[3].filter(ImageFilter.GaussianBlur(4))

fig3 = plt.figure()
rows=1
columns=2

fig3.add_subplot(rows,columns,1)
plt.imshow(imgs[3])
plt.axis("off")
plt.title("Before")

fig3.add_subplot(rows,columns,2)
plt.imshow(blurred)
plt.axis("off")
plt.title("After")
```

```python
#sa se identifice muchiile ontr-o imagine si sa se afiseze in format "before-after"
muchii = imgs[6].convert("L")
muchii = muchii.filter(ImageFilter.FIND_EDGES)

fig4 = plt.figure()
rows=1
columns=2

fig4.add_subplot(rows,columns,1)
plt.imshow(imgs[6])
plt.axis("off")
plt.title("Before")

fig4.add_subplot(rows,columns,2)
plt.imshow(muchii, cmap='gray')
plt.axis("off")
plt.title("After")
```
