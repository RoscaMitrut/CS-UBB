```python
'''
propozitie (string): propozitia a carei cuvant "maxim" dorim sa il aflam
return (string): cuvantul "maxim"
Complexitate: O(n log n)

Functia desparte propozitia in cuvinte, le sorteaza, si il ia pe ultimul ("cel mai mare")
'''
def rez1_1(propozitie):
	lista_cuvinte = propozitie.split()
	lista_cuvinte.sort()
	return lista_cuvinte[-1]

'''
propozitie (string): propozitia a carei cuvant "maxim" dorim sa il aflam
return (string): cuvantul "maxim"
Complexitate: O(n)

Funcita desparte propozitia in cuvinte, initializeaza o variabila cu primul cuvant din propozitie,
si o compara cu fiecare cuvant din propozitie, actualizand-o de fiecare data cand cuvantul la care suntem este "mai mare"
In final, vom avea in variabila "actual" cuvantul "maxim"
'''
def rez1_2(propozitie):
	lista_cuvinte = propozitie.split()
	actual = lista_cuvinte[0]
	for el in lista_cuvinte:
		if(el[0]>actual):
			actual = el
	return actual


assert(rez1_1("Ana are mere si pere")=="si")
assert(rez1_2("Ana are mere si pere")=="si")
assert(rez1_1("the quick fox jumps over the lazy dog")=="the")
assert(rez1_2("the quick fox jumps over the lazy dog")=="the")
```

```python
'''
coord1 (pair de real): coordonatele primului punct
coord2 (pair de real): coordonatele celui de al doilea punct
return (real): distanta dintre cele 2 puncte
Complexitate: O(1)

Calculam distanta folosind formula
'''
def rez2_1(coord1,coord2):
	return math.sqrt( ((coord2[0]-coord1[0])**2) + ((coord1[1]-coord2[1])**2) )

'''
coord1 (pair de real): coordonatele primului punct
coord2 (pair de real): coordonatele celui de al doilea punct
return (real): distanta dintre cele 2 puncte
Complexitate: O(1)

Transformam coordonatele in coordonate polare si calculam distanta folosind formula
'''
def rez2_2(coord1,coord2):
	r1 = math.sqrt(coord1[0]**2 + coord1[1]**2) #distance
	a1 = math.atan(coord1[1]/coord1[0]) #angle
	r2 = math.sqrt(coord2[0]**2 + coord2[1]**2)
	a2 = math.atan(coord2[1]/coord2[0])
	return math.sqrt(r1**2 + r2**2 - 2*r1*r2*math.cos(a1-a2))

assert(rez2_1((1,5),(4,1))==5.0)
assert(rez2_2((1,5),(4,1))==5.0)
```

```python
'''
cuvinte (string): string de cuvinte in care cautam string-urile care apar o singura data
return (list de string-uri): string-urile care apar o singura data in string-ul initial
Complexitate: O(n+m)

Functia desparte propozitia in cuvinte si creeaza un dictionar de frecventa, dupa care returnam o lista cu cuvintele care au valoarea 1 (frecventa 1) in dictionar
'''
def rez4_1(cuvinte):
	lista_cuvinte = cuvinte.split()
	frecventa = {}
	for el in lista_cuvinte:
		if el in frecventa:
			frecventa[el]+=1
		else:
			frecventa[el]=1
	lista_finala = []
	for key,value in frecventa.items():
		if(value==1):
			lista_finala.append(key)
	return lista_finala

'''
cuvinte (string): string de cuvinte in care cautam string-urile care apar o singura data
return (list de string-uri): string-urile care apar o singura data in string-ul initial
Complexitate: O(n^2)

Functia verifica fiecare cuvant din lista, cu fiecare cuvant din lista(fara el insusi) si in cazul in care nu e duplicat, e adaugat in lista finala
'''
def rez4_2(cuvinte):
	lista_cuvinte = cuvinte.split()
	lista_finala = []
	for i in range(len(lista_cuvinte)):
		switch = False
		for j in range(len(lista_cuvinte)):
			if(lista_cuvinte[i]==lista_cuvinte[j] and i!=j):
				switch=True
		if(switch==False):
			lista_finala.append(lista_cuvinte[i])
	return lista_finala

assert(rez4_1("ana are ana are mere rosii ana")==["mere","rosii"])
assert(rez4_2("ana are ana are mere rosii ana")==["mere","rosii"])
assert(rez4_1("a a a a b c c c c d e e e e")==["b","d"])
assert(rez4_2("a a a a b c c c c d e e e e")==["b","d"])
```

```python
'''
numere (lista de numere): sirul in care cautam valoarea duplicata
return (int): valoarea care se repeta
Complexitate: O(n)

Functia returneaza diferenta dintre suma sirului si suma sirului (cu lungime mai mica cu 1) daca acesta nu ar avea un element duplicat
'''
def rez5_1(numere):
	suma = sum(numere)
	lungime = len(numere)
	suma_fara_duplicata = lungime * (lungime-1)/2
	return suma-suma_fara_duplicata

'''
numere (lista de numere): sirul in care cautam valoarea duplicata
return (int): valoarea care se repeta
Complexitate: O(n log n)

Functia sorteaza sirul, dupa care trecem prin el, verificand daca avem 2 elemente egale consecutive
'''
def rez5_2(numere):
	numere.sort()
	for i in range(len(numere)-1):
		if(numere[i]==numere[i+1]):
			return numere[i]

'''
numere (lista de numere): sirul in care cautam valoarea duplicata
return (int): valoarea care se repeta
Complexitate: O(n^2)

Functia verifica fiecare element din lista, cu fiecare element din lista(fara el insusi) si in cazul in care e duplicat, e returnat
'''
def rez5_3(numere):
	for i in range(len(numere)):
		for j in range(len(numere)):
			if(numere[i]==numere[j] and i!=j):
				return numere[i]

assert(rez5_1([1,2,3,4,5,3])==3)
assert(rez5_2([1,2,3,4,5,3])==3)
assert(rez5_3([1,2,3,4,5,3])==3)
```

```python
'''
numere (lista de numere): sirul in care cautam elementul majoritar
return (int): elementul majoritar
Complexitate: O(k) + O((n - k) log k) + O(k log k)
            k=1 aici (deoarece avem nevoie de un singur element, nu de frecventa tuturor elementelor)

Functia foloseste metoda most_common() pentru a return elementul majoritar in lista
'''
def rez6_1(numere):
	return Counter(numere).most_common(1)[0][0]

assert(rez6_1([1,2,3,4,1,1])==1)
```

```python
'''
numere (lista de numere): sirul in care cautam al k-lea cel mai mare element
k (int): al k-lea cel mai mare element pe care il cautam
Complexitate: O(n log n)

Functia recreeaza lista fara duplicate, dupa care o sorteaza si returneaza elementul de pe pozitia k
'''

def rez7_1(numere,k):
	fara_dup = []
	[fara_dup.append(x) for x in numere if x not in fara_dup]
	fara_dup.sort(reverse=True)
	return fara_dup[k-1]

assert(rez7_1([9,8,7,6,5,4],2)==8)
```
